```python
import snscrape.modules
import pandas as pd

keywords = ["حرب روسيا و أوكرانيا","روسيا و أوكرانيا","بوتين و أوكرانيا","أزمة أوكرانياو روسيا"]

tweets_list = []
for keyword in keywords: 
    tweets = snscrape.modules.twitter.TwitterSearchScraper(keyword + 'since:2022-08-01 until:2022-08-02 lang:ar')
    
    for tweet in tweets.get_items():
        print(tweet)
        if len(tweets_list) < 200:
            break
        tweets_list.append(tweet.content)
        
tweet_df = pd.DataFrame(tweets_list, columns=["Text"])
tweet_df.to_csv('test١.csv',index=False,encoding='utf-8')
```
